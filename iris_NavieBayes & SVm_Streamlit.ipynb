```python
x=df.iloc[:,1].values
y=df.iloc[:,0].values
```

```python
x[0]
```

```python
x[1]
```

```python
x[2]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(stop_words='english')
```

```python
x_train_vect=vect.fit_transform(x_train)
x_test_vect=vect.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model1=MultinomialNB()
```

```python
model1.fit(x_train_vect,y_train)
```

```python
y_pred1 = model1.predict(x_test_vect)
y_pred1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
sample1=['win free tickets']
sample1=vect.transform(sample1)
model1.predict(sample1)
```

```python
sample1=['hi jo']
sample1=vect.transform(sample1)
model1.predict(sample1)
```

```python
# using NB pipeline
from sklearn.pipeline import make_pipeline
model2=make_pipeline(CountVectorizer(),MultinomialNB())
model2.fit(x_train,y_train)
```

```python
str1=["hello and welcome to the intel program","free free free"]
model2.predict(str1)
```

```python
y_pred2 = model2.predict(x_test)
y_pred2
```

```python
model2.predict(["free free free"])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred2)
```

```python
#using svm for model building
from sklearn.svm import SVC
model3=SVC().fit(x_train_vect,y_train)
```

```python
y_pred3=model3.predict(x_test_vect)
y_pred3
```

```python

op=vect.transform(['sweet home'])
model3.predict(op)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred3)
```

```python
#using pipeline for SVM 
from sklearn.pipeline import make_pipeline
model4=make_pipeline(CountVectorizer(),SVC())
model4.fit(x_train,y_train)
```

```python
y_pred4 = model4.predict(x_test)
y_pred4
```

```python
model4.predict(["free free free"])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred4)
```
